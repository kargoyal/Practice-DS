```python
def die2(x=36):
    co=0
    for i in range(x):
        di1=random.randint(1,6)
        di2=random.randint(1,6)
        sc=di1+di2
        if sc > 7 or sc %2==0:
            co+=1
    return ([co,di1,di2])
    
```

```python
pp={}
for i in range(36):
    
    pp[(die2()[0]/36)*100]=die2()[1],die2()[2]
```
